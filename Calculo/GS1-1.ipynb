```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2003, 2005, 2009, 2010, 2016, 2019]
y = [27.7, 29.3, 34.4,35.6, 44.3, 55]

plt.plot(x, y, color = "blue", label="Casos de demência")
plt.scatter(x, y, color='red')

plt.xlabel('Ano')
plt.ylabel('Casos em milhões')

plt.title('Aumento dos casos de demência')

plt.legend()
plt.grid()

plt.show()
```

```python
x = [2003, 2005, 2009, 2010, 2016, 2019]
y = [156, 315.4, 422, 604, 948, 1300]

plt.plot(x, y, color = "blue", label="Custo com demência")
plt.scatter(x, y, color='red')

plt.xlabel('Ano')
plt.ylabel('Custo em bilhões (US$)')

plt.title('Aumento do custo mundial com demência')

plt.legend()
plt.grid()

plt.show()
```

```python
x = [2003, 2005, 2009, 2010, 2016, 2019]
y = [.877, .946, 1.110, 1.160, 1.450, 1.620]

plt.plot(x, y, color = "blue", label="Mortes por demência")
plt.scatter(x, y, color='red')

plt.xlabel('Ano')
plt.ylabel('Mortes em milhões')

plt.title('Mortes por demência')

plt.legend()
plt.grid()

plt.show()
```
